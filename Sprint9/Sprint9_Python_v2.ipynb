1. Calculadora del √≠ndice de masa corporal

Escribe una funci√≥n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber√° ingresar estos datos. Puedes obtener m√°s informaci√≥n de su c√°lculo en: üîó √çndice de masa corporal IMC que es y c√≥mo se calcula.

La funci√≥n debe clasificar el resultado en sus respectivas categor√≠as.

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado o no toma valores razonables.

```python
#Python program to calculate the BMI (or IMC in Spanish)

# BMI calculation function
def calc_bmi (weight:float, high:float) -> float:
    """
    Do the calculation of the BMI.
    value: float
    """
    return weight / (high ** 2)

# BMI classification function
def bmi_classification(calc: float) -> str:
    """
    Classify the BMI according to the BMI calculation
    value: float
    """

    if calc < 18.5:
        return 'Low Weight'
    elif calc <= 24.99:
        return 'Normal Weight'
    elif calc <= 29.99:
        return 'Over Weight'
    else:
       return'Obesity'

def get_data(prompt: str) -> str:
    """
    Get the user input
    value: str
    """
    return input(prompt)

# function for data input validation. 
def validate_input(value: str, name: str, min_value: float, max_value:float) -> float:
    """ 
    Validade if the input data is float.
    value:float
    """
    try:
        value = float(value)
        if value <= 0:
            print(f'The {name} must be greater than zero. Please try again.')
            return None
        if value < min_value or value > max_value:
            print(f'Unrealistic value. The {name} must be between {min_value} and {max_value}')
            return None
        return value
    except ValueError:
        print(f"Invalid input for {name}. Please enter a number.")
        return None
    
# main function
def bmi():
    """
    Main function. Calculates and classify the BMI
    value: None
    """
    while True:
        weight_imput = get_data('Enter your WEIGHT in kilograms (Kg):')
        weight = validate_input(weight_imput, 'WEIGHT', 20, 400)
        
        if not weight:
            continue

        high_input = get_data('Enter your HIGH in meters (m):')
        high = validate_input(high_input, 'HIGH', 0.5, 3.0)

        if not high:
            continue

        
        bmi_value = calc_bmi(weight, high)
        bmi_class = bmi_classification(bmi_value)
        display_result(weight, high, bmi_value, bmi_class)
        break

# Display the data in the terminal
def display_result(weight: float, high: float, bmi_value: float, bmi_class: str ) -> None:
    """ 
    Display the BMI value and classification.
    value: None
    """
    print(f'Your WEIGHT is: {weight} Kg')
    print(f'Your HIGH is: {high} m')
    print(f'Your BMI is {bmi_value:.2f} - {bmi_class}')

# Main function call
bmi()
```

2. Conversor de temperaturas

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las m√°s comunes son Celsius (¬∞C), Fahrenheit (¬∞F) y Kelvin (K). Tambi√©n existen otras unidades como Rankine (¬∞Ra) y R√©aumur (¬∞Re).

Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como m√≠nimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado.

F√≥rmula	(0 ¬∞C √ó 9/5) + 32 = 32 ¬∞F -> (9/5 = 1.8)

F√≥rmula	0 ¬∞C + 273,15 = 273,15 K

F√≥rmula	(0 ¬∞F ‚àí 32) √ó 5/9 = -17,78 ¬∞C 

F√≥rmula	(0 ¬∞F ‚àí 32) √ó 5/9 + 273,15 = 255,372 K

F√≥rmula	0 K ‚àí 273,15 = -273,1 ¬∞C

F√≥rmula	(0 K ‚àí 273,15) √ó 9/5 + 32 = -459,7 ¬∞F


```python
# Python program to convert a temperature in another 2 different units

# function for getting the temperature unit
def get_unit() -> str:
 
    """ 
    validate the input data  
    value: str

    """
    # validates unit
    while True:
        temp_unit = input('Please choose the temperature unit to convert: For "CELSIUS" press "C", for "FAHRENHEIT" press "F", for "KELVIN" press "K"').upper()

        if temp_unit in ['C', 'F', 'K']:
            break 
        else:
            print('Invalid choice. Please enter C, F or K') 

    return temp_unit  

# function for getting the temperature for converting
def get_temperature(temp_unit) -> float:
    """ 
    get and validades temperatura input
    value: float
    """
  
    while True:
        try:
            temperature = float(input(f'Enter the temperature in {temp_unit}:'))

            if temp_unit == 'K' and temperature < 0:
                print('Kelvin cannot be negative. Please enter a valid Kelvin temperature')
                continue
            break
        except ValueError:
            print('Invalid input, please enter a numeric value')

    return temperature


# function to perform the temperature conversion
def temperature_converter(temp_unit, temperature) -> None:

    """ 
    execute the temperature unit conversion 
    value: None

    """
    # declare a dictionary to store the temperature units 
    temp_converter_dict = {}
    
    # perform the calcuation for converting and store the unit names
    match temp_unit:
        case 'C':
            conversion1 = (temperature * 1.8) + 32
            conversion2 = temperature + 273.15
            name1 = '¬∞F'
            name2 = '¬∞K'
                 
        case 'F':
            conversion1 = (temperature - 32) * (5/9)
            conversion2 = conversion1 + 273.15
            name1 = '¬∞C'
            name2 = '¬∞K'
                   
        case 'K':
            conversion1 = temperature - 273.15
            conversion2 = (conversion1 * 1.8) + 32
            name1 = '¬∞C'
            name2 = '¬∞F'
           
    # store the original unit data and the converted temperatures in a dictionary        
    temp_converter_dict[temperature] = {
    'original' : temp_unit, 
    'converted' : { 
        name1: round(conversion1, 2), 
        name2: round(conversion2, 2) 
    }
}
    
    return temp_converter_dict

# main function
def converter ():
    """
    Main function. Reads unit, temperature and converts the temperature
    """
    unit = get_unit() # get temperature unit
    temperature = get_temperature(unit) # get the temperature             
    converted_temperature = temperature_converter(unit, temperature) # do temperature conversion
    display_result(converted_temperature) # display the result


def display_result(converted_temperature) -> None:
    """
    print the temperature conversion result
    value: None
    """
     
    for temp, data in converted_temperature.items():
        print(f"{temp} ¬∞{data['original']} ‚Üí ", end="")
        for unit, value in data['converted'].items():
            print(f"{value} {unit}", end="  ")
        print()

# function call   
converter()
```

3. Contador de palabras de un texto.

Escribe una funci√≥n que dado un texto, muestre las veces que aparece cada palabra. Intenta que se gestionen todas las posibles casu√≠sticas que hagan que el programa no funcione correctamente.

(EXTRA): ¬øCu√°l es la longitud media de las palabras del texto que has escrito? ‚Äú¬øHola c√≥mo va?‚Äù deber√≠a devolver (4+3+2) / 3 = 3

```python
import re
from collections import Counter
from statistics import mean
from pprint import pprint

# function that receive the text imput for counting
def read_text() -> str:
        """
        receive the text input
        value: str
        """
        return input('Enter the text for words counting') 
        

# function that transform the text input in a list of words
def generate_list(text: str) -> list:
        """ 
        generates a list of words based on the text input
        value: list
        """
        return re.findall(r'\b\w+\b', text.lower()) # \b matches the whole word and ignores punctuation


# function for counting the words
def count_words(list_words: list) -> dict: 
        """ 
        returns a dictionary of the words and the counting and the list of words to be used for the average calculation function        
        value: dict
        """
        return dict(Counter(list_words)) # dictionary stores the words as keys and the counting as values
        


# function that counts the total of letters in each word of the given text and return the average.
def calculate_avg_word_length(dict_count_words) -> dict:
        """
        calculate the average length of the words
        value: dict
        """
        return mean(map(len, dict_count_words.keys()))

        
# function that prints the list of words and the total of each word
def display_list_words(dict_count_words) -> dict:
       """ 
       display the dictionary with the list of words and the couting
       value: dict
       """
       pprint(f'Here is the list of words and the total of occurences: {dict_count_words}')

# prints the average length of
def display_average_words(average_words_length) -> float: 
        """ 
        display the average word lenght
        value: float
        """
        print(f'The average word length is: {average_words_length:.2f}')

# orchestrator function
def main():
        """ 
        orchestrator function
        value: None
        """
        text = read_text()
        word_list = generate_list(text)
        dict_words = count_words(word_list)
        average_words = calculate_avg_word_length(dict_words)
        display_list_words(dict_words)
        display_average_words(average_words)

# main function call
main()
```

4. Diccionario inverso (con posibilidad de duplicados)

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiando las claves y valores. Los valores y claves en el diccionario original son √∫nicos; si √©ste no es el caso, la funci√≥n deber√≠a imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.

```python
from collections import defaultdict

# fuction that returns a reverse dictionary even when there are duplicates
def reversing_dicctionary (original_dicct) -> dict:
    """ 
    reverse the original dictionary
    value: dict
    """
    reverse_dicct = defaultdict(list)

    for key, value in original_dicct.items():
        reverse_dicct[value].append(key)

    final_dict = {}
    for value, keys in reverse_dicct.items():
        if len(keys) == 1:
            final_dict[value] = keys[0]
        else:
            final_dict[value] = keys


    return final_dict

# function for checking duplicates
def check_duplicates (reverse_dicct) -> dict:
    """ 
    check duplicates in the dicctionary
    value: dict
    """
    for value, keys in reverse_dicct.items():
        if isinstance(keys, list) and len(keys) > 1:
            print(f"Error: Duplicated value '{value}' found for keys {keys}") 
         

# main function for reversing dictionary and display the result
def main(original_dicct):
    """
    orchestrator function
    value: none    
    """
    reversed_dict = reversing_dicctionary (original_dicct)
    print(reversed_dict)
    check_duplicates(reversed_dict)

#test
original_dicct = {'x': 'apple', 'y': 'banana', 'z': 'banana'} # test dicctionary
main(original_dicct) #function call
```

1. Contador y ordenador de palabras de un texto.

El cliente qued√≥ contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario seg√∫n la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a Z. "tu_me_quieres_blanca.txt" la salida esperada ser√≠a:

```python
import re
from collections import Counter, defaultdict
import locale
from pprint import pprint

def read_file() -> str:
    """ 
    read a .txt file.
    value: str
    """
    with open('tu me quieres blanca.txt', encoding = 'utf-8') as sentence:
        return sentence.read()

        
def generate_list(text: str) -> list:
        """ 
        generate a list of the words in the .txt file
        value: list
        """
        return re.findall(r'\b\w+\b', text.lower()) # \b matches the whole word and ignores punctuation

def count_words(list_words: list) -> dict: 
        """ 
        return a dictionary of the words and the counting 
        value: dict        
        """
        return dict(Counter(list_words)) # dictionary stores the words as keys and the counting as values


# function for counting the words in a givem txt file and order them alphabetically
def group_words(count_words) -> dict:
    """  
    group by first letter and sort alphabetically
    value: dict
    """
    #group words by first letter
    grouped_words = defaultdict(dict)
    for word, count in count_words.items():
        grouped_words[word[0]][word] = count

    locale.setlocale(locale.LC_ALL, 'es_ES.UTF-8') # for proper alphabetical sorting
    
    # sort words alphabetically in the groups
    for letter in grouped_words:
        grouped_words[letter] = dict(sorted(grouped_words[letter].items(), key=lambda x: locale.strxfrm(x[0])))

    # variable that stores the sorted dictionary
    ordered_dict = dict(sorted(grouped_words.items(), key=lambda x: locale.strxfrm(x[0])))

    # function returns a dictionary sorted alphabetically by the keys        
    return ordered_dict

# display function
def result_output(data) -> None:
     """  
     display the sorted dictionary
     value: None
     """
     pprint(data)

# main function
def count_and_sort_words() -> None:
    """ 
    orchestrator function
    value: None
    """
    text_file = read_file()
    list_words = generate_list(text_file)
    words_counting = count_words(list_words) 
    words_grouping = group_words(words_counting)
    result_output(words_grouping)

# main function call
count_and_sort_words()
```

2. Conversi√≥n de tipos de datos.

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estar√°n todos los elementos que pudieron convertirse en flotantes y la otra donde est√°n los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente: 
[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python

# function that returns a list of elements converted to float and another list of the elements impossible to convert to float.
def extract_numbers(customer_list) -> list:
    """ 
    return a list of elements converted to float and a list with the ones that cannot be converted
    value: list
    """
     
    float_list = []
    not_float_list = []
    stack = [customer_list] # to store the elements from the variable customer_list that are lists or tuples 

    while stack:
        current = stack.pop() # takes the first element form stack
        
        for element in current:
            if isinstance (element, (list, tuple)):
                stack.append(element)  #append the element to stack list if the element is a list or a tuple
            else:
                try:
                    float_list.append(float(element)) # if the element is converted to float append to the float_list

                except (ValueError, TypeError):
                    not_float_list.append(element) # if the element connot be converted to float then append it to the not_float_list

    return float_list, not_float_list

# Test

customer_list = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

# function call and return stores to the objects float_values and non_float_values
float_values, non_float_values = extract_numbers(customer_list)

# prints the list with the float values and the non-float values
print(f'Values converted to float: {float_values}')
print(f'Values not converted to float: {non_float_values}')
```
