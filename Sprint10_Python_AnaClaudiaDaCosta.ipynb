Utiliza la función que acabas de crear para generar una columna nueva en DataFrame con la edad actual.

```python
# create a column 'Fecha' and rename columns to be used as date_Time format
df_sorted['Fecha_Nacimiento'] = pd.to_datetime(df_sorted[['Dia', 'Mes', 'Año']].rename(columns={'Dia':'day', 'Mes':'month', 'Año':'year'}))
```

```python
# apply the date_time configuration
df_sorted['Fecha_Nacimiento'] = pd.to_datetime(
    df_sorted['Dia'].astype(str) + '-' +
    df_sorted['Mes'].astype(str) + '-' +
    df_sorted['Año'].astype(str),
    format='%d-%m-%Y'
)
```

```python
# function that calculates the current age
def current_age(birth_date):

    now = datetime.now()
    age = now.year - birth_date.year

    if (now.month, now.day) < (birth_date.month, birth_date.day):
        age -= 1    
    return age

# 'Edad_Actual' column creation and currente_age function call
df_sorted['Edad_Actual'] = df_sorted['Fecha_Nacimiento'].apply(current_age) 
```

```python
# show info on 'iris' DataSet
df_iris.info()
```

```python
# function definition
def create_export_graphs(data_frame) -> None:
    """ 
    creates an histogram for each numeric column from a DataFrame
    value: None
    """

    # numeric columns
    numeric_graph = data_frame.select_dtypes(include='number') # select only the type number variables
    
    if not numeric_graph.empty:
        
        for columns in numeric_graph.columns:

            plt.figure()
            numeric_graph[columns].hist()
            plt.title(f'Distribution on column {columns}')
            plt.savefig(f'Histogram on column {columns}')

    
    # datetime columns
    datetime_graph = data_frame.select_dtypes(include='datetime')
    
    if not datetime_graph.empty:

        for columns in datetime_graph.columns:
            plt.figure()
            years = datetime_graph[columns].dt.year
            years.value_counts().sort_index().plot(kind='bar')
            plt.title(f'Most frequent years: {columns}')
            plt.savefig(f'Years_{columns}')


    # categorical columns
    categorical_graph = data_frame.select_dtypes(include=['object', 'category'])

    if not categorical_graph.empty:
        
        for columns in categorical_graph.columns:
            plt.figure()
            categorical_graph[columns].value_counts().head(10).plot(kind='bar')
            plt.title(f'Most frequent values: {columns}')
            plt.xticks(rotation=45)
            plt.savefig(f'Categories_{columns}')

    plt.show()
        
           
         
```

```python
# function call
create_export_graphs(df_iris)
```

```python
# function for propose minimum distance route taking the city input as start point.
def propose_route(data_frame):
    city = input(f'Choose the city to start the route:')
    route = [city] # to store the list of cities in the route
    total_distance = 0 # to calculate the total distance

    data_frame_copy = data_frame.copy() # copy of the data_frame

    while len(data_frame_copy) > 1: # loop for creating the city list

        next_city = data_frame_copy[city].idxmin() # takes the nearest distance based on the minimum value
        distance = data_frame_copy.loc[next_city, city] # takes the related value
    
        route.append(next_city) # update the route list with the next city
        total_distance += distance # increments the related distance
    
        data_frame_copy.drop(city, axis=1, inplace=True) # drop the city from the matrix column
        data_frame_copy.drop(city, axis=0, inplace=True) # drop the city from the matrix index
    
        city = next_city #takes to next city value to iterate

    return print(f'The rout proposed for starting the trip from {route[0]} is:\n{route}.\nThe total distance is {total_distance} Km')
  
```

```python
# function call
propose_route(df_rutas)
```
