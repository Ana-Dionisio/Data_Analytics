2. Muestra por pantalla:

```python
# function for retrieving the quantity and the request status code

def get_api_quantity(list):
    
    url = 'https://jsonplaceholder.typicode.com' + list

    response = requests.get(url)

    if response:
        data = response.json()
        message = print(f'Total {list}: {len(data)} and Status code: {response.status_code}')
        #pprint(f'Status code {list}: {response.status_code}')
    else:
        message = print(f'Error retrieving data from "{list}". Status code: {response.status_code}')
    
    return message   
    
```

```python
# function call

get_api_quantity('/posts')
get_api_quantity('/users')
get_api_quantity('/todos')
```

4. Realiza una petición POST para crear una nueva publicación ficticia. Incluye el título, el cuerpo del mensaje y un userId.

Muestra la respuesta JSON.
Muestra el código de estado.

```python
# function for creating a API posting

def post_data(mypost):
        url = 'https://jsonplaceholder.typicode.com/posts'

        response = requests.post(url, json = mypost)

        if response:
               message = print(f'Success posting data! Status code: {response.status_code}')
                
        else:
                message = print(f'Error posting data to {url}". Status code: {response.status_code}')
        
        return message, response.json()
```

```python
# object and funtion call

mypost = {"userId":600,
      "title":"title test sprint 12 data analytics",
      "body":"body test sprint 12 data analytics"}

post_data(mypost)
```

5. Realiza una petición PATCH para modificar parcialmente una publicación existente.

Muestra la respuesta JSON.  Muestra el código de estado.

```python
# function for patching a API posting

def pathc_data(post_id, myupdate):
        
        url = 'https://jsonplaceholder.typicode.com/posts/' + post_id

        response = requests.patch(url, json = myupdate)

        if response:
               message = print(f'Success patching data! Status code: {response.status_code}')
                
        else:
                message = print(f'Error pathing data to {url}". Status code: {response.status_code}')
        
        return message, response.json()
```

```python
# object and function call 

mypost = {'title': 'Updated title via PATCH'}
id = '501'
pathc_data(id, mypost)
```

6. Haz una petición DELETE sobre una publicación.

Muestra la respuesta JSON.
Muestra el código de estado.

```python
# function for deleting  a API posting

def delete_data(post_id):
        
        url = 'https://jsonplaceholder.typicode.com/posts/' + post_id

        response = requests.delete(url)

        if response:
               message = print(f'Success deleting data! Status code: {response.status_code}')
                
        else:
                message = print(f'Error deleting data. Post id {post_id}". Status code: {response.status_code}')
        
        return message, response.json()
```

```python
# object and function call

id = '501'
delete_data(id)
```

1. Utiliza la API de Open Data BCN para:  

Buscar un dataset de tu interés mediante package_search y package_show. .

```python
# function for package_search

def search_dataset(texto_busqueda):
    search = url + 'package_search'
    params = {'q': texto_busqueda}

    response = requests.get(search, params=params)
    data = response.json()

    if response.ok and data['success']:
        result= data['result']['results']
        return result
    else:
        raise Exception('Error en package_search')
```

```python
# object and function call

url = 'https://opendata-ajuntament.barcelona.cat/data/api/3/action/'

result = search_dataset('bicicleta')
print(f'Datasets founds: {len(result)}')
for r in result[:5]:
    print('-', r['title'])
```

```python
# function for package_show

def show_dataset(id):
    search = url + 'package_show'
    params = {'id': id}

    response = requests.get(search, params=params)
    data = response.json()

    if response.ok and data['success']:
        return data['result']
    else:
        raise Exception('Error en package_show')
```

```python
# object and function call

selection = result[0]
detail = show_dataset(selection['id'])

print('Title:', detail['title'])
print('Description', detail['notes'])
print('Resources availables:')
for resource in detail['resources']:
    print('-', resource['name'], resource['format'], resource['url'])
```
